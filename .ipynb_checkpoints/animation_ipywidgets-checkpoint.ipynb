# Image Sequences with PIL and ipywidgets

This is a quick demo of using [PIL/pillow](https://python-pillow.org/) to display images in a Jupyter notebook in a lightweight manner, alon with [ipywidgets](https://ipywidgets.readthedocs.io/) to scan through a set of images interactively.

To install the packages used here, use either

```
$ conda install numpy pillow ipywidgets
```

or

```
$ pip install numpy pillow ipywidgets
```

````python
import numpy as np
from PIL import Image

def display_image(x):
    x_scaled = np.uint8(255 * (x - x.min()) / x.ptp())
    return Image.fromarray(x_scaled)

display_image(np.random.rand(200, 200))
````

````python
from ipywidgets import interact

def display_sequence(images):
    def _show(frame=(0, len(images)-1)):
        return display_image(images[frame])
    return interact(_show)

# Make some images
images = [np.random.rand(200, 200) for i in range(150)]
for i, image in enumerate(images):
    image[i:i+50, i:i+50] = 0

# Note: this will only display with a live kernel
display_sequence(images);
````

Note: if you'd like to save the images to an animated gif directly, the [imageio](http://imageio.github.io) package gives you an easy route:

````python
import imageio
images_scaled = [np.uint8(255 * image) for image in images]
imageio.mimsave('movie.gif', images_scaled, format='gif', fps=60)
````
